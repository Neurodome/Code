```python
import yt.imods
import numpy
import NeurodomeLibrary as NL
zoomfactor = 0.45
```

```python
pngdir = '/Users/subbaraojeffrey/Documents/MATLAB/Neurodome/'
data = numpy.zeros([128*2,128,128*2])
data[:128,:,:128] = NL.loadtiff(pngdir+'40_31_10.tif')
data[:128,:,128:] = NL.loadtiff(pngdir+'40_31_11.tif')
data[128:,:,:128] = NL.loadtiff(pngdir+'40_32_10.tif')
data[128:,:,128:] = NL.loadtiff(pngdir+'40_32_11.tif')
pf,tf = NL.ytload(data,bbox=numpy.array([[-3,3],[-1.5,1.5],[-3,3]]))
```

```python
c = (pf.domain_right_edge + pf.domain_left_edge)/2.0
L = np.array([0.,1.,0.])
W = 1.8
N = 400
cam = pf.h.camera(c, L, W, N, tf, fields = ["Density"], log_fields = [False])
cam.zoom(zoomfactor)
```

```python
#cam.zoom(.75)
cam.show(clip_ratio=1)
```

```python
minim,maxim = pf.h.all_data().quantities["Extrema"]('Density')[0]
```

```python
for val in numpy.unique(data)[1:]:
    tf.clear()
    tf.add_gaussian(val,.1,[1.,1.,0,1.])
    cam.snapshot('EM_cubes_at'+str(val),clip_ratio=1.5)
```

```python
for i,theta in enumerate(numpy.linspace(0,numpy.pi,6,endpoint=False)):
    cam.switch_view(normal_vector=[numpy.cos(theta),0.,numpy.sin(theta)],north_vector=[0.,1.,0.])
    cam.snapshot("Neurons_Mock"+"0"*(2-len(str(i)))+str(i)+".png",clip_ratio=1.)
```

```python
#cam.zoom(1./20)
cam.show(clip_ratio=3)
```

```python
for i in xrange(int(minim),int(maxim)+1):
    tf.clear()
    tf.add_gaussian(float(i),1.,[1.,1.,0.,1.])
    cam.snapshot("TF_scan"+"0"*(3-len(str(i)))+str(i)+".png",clip_ratio=4.)
```
