```python
import yt.imods
import numpy
import NeurodomeLibrary as NL
zoomfactor = 0.432
```

```python
pngdir = '/Users/subbaraojeffrey/Documents/MATLAB/Neurodome/small stack 1/'
pngname = 'smoothed_image_*.png'
pf,tf = NL.ytload(NL.loadpngseries(pngdir+pngname))
```

```python
c = (pf.domain_right_edge + pf.domain_left_edge)/2.0
L = np.array([0.,1.,0.])
W = 1.8
N = 200
cam = pf.h.camera(c, L, W, N, tf, fields = ["Density"], log_fields = [False])
cam.zoom(zoomfactor)
```

```python
cam.show(clip_ratio=4.)
```

```python
minim,maxim = pf.h.all_data().quantities["Extrema"]('Density')[0]
```

```python
tf.clear
tf.add_gaussian(66.,4.,[0.,.2,1.,1.])
tf.add_gaussian(88.,3.,[1.,.5,0.,.8])
```

```python
for i,theta in enumerate(numpy.linspace(0,numpy.pi,100)):
    cam.switch_view(normal_vector=[numpy.cos(theta),0.,numpy.sin(theta)],north_vector=[0.,1.,0.])
    cam.snapshot("Serial_EM_Mock"+"0"*(2-len(str(i)))+str(i)+".png",clip_ratio=4.)
```

```python
cam.rotate(numpy.pi/2,L)
cam.show(clip_ratio=1.)
```

```python
for i in xrange(int(minim),int(maxim)+1):
    tf.clear()
    tf.add_gaussian(float(i),1.,[1.,1.,0.,1.])
    cam.snapshot("TF_scan"+"0"*(3-len(str(i)))+str(i)+".png",clip_ratio=4.)
```
